1.	Общее число оценок в файле
2.	Общее количество пользователей, поставивших оценки
3.	Общее количество оцененных фильмов
4.	ID самого активного пользователя
5.	Фильм, собравший наибольшее количество оценок


### Цель задач: научится работать с большим объемом данных и форматом csv.

### Решение задачи: 

- Для подсчета общего количества оценок использовал num_of_rating() где посчитал длину столбца с оценками. В случае если будет NaN значения исполнение будет неверным, но пропусков нету, поэтому оставил так.

- Для количества пользователей и фильмов тоже самое за исключением того что обеднил их по уникальному id с помощью unique(). Так же будет неправильно исполняться в случае если фильм никто не оценил или пользователь не поставят оценку, опять же здесь все поставили оценки и всем фильмам.

- Для подсчета самого активного пользователя и фильма с наибольшим количеством оценок использовал most_active_userId() и most_rated_film(). Тут группировал по нужному столбцу и сортировал по количеству записей в других столбцах брав первый элемент с конца(с наибольшим количеством оценок) и доставал оттуда id с помощью .index и само значение с помощью .values.
