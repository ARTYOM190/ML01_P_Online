Задание: Напишите программу, которая рассчитывает расход топлива маршрута, на вход подается длина маршрута и расход топлива на 100 километров. Используйте try catch (проверка что пользователь ввел число>0).

Просто перемножаем расстояние на расход и получаем ответ, единственное сделал отдельные функции для инпута значений, что бы в ник проверять корректность ввода.
В ходе выполнения решил по экспериментировать со словарем, в котором записаны меры исчисления расстояния и коэффициент перевода. Программа делит строку пробелом, достает значение по ключу и умножается на коэффициент. Так же можно использовать регулярные выражения, но данная для данной конструкции ввода (ЧИСЛО МЕРА) мне кажется она не нужна. Еще нужно реализовать проверку числа на 0 и так как мы используем .split проверку на ошибку индекса массива.
Буду доделывать.

Сделал проверку для второго значения.

```python
dist_calc_measures = {
    "км": 0.01,
    "м": 0.001,
    "мили": 1.609344
}

def fuel_input():
    while(True):
        fuel = float(input('Введите колличество топлива: '))
        try:
            if fuel <= 0:
                raise ValueError
            else:
                return fuel
        except ValueError:
            print("Значенте введено неверно, поробуйте снова")
        

def length_of_route_input():
    while(True):
        length_of_route = input('Введите длинну маршрута: ')
        try:
            length_of_route = float(length_of_route.split(' ')[0]) * dist_calc_measures[length_of_route.split(' ')[1]]
            if  length_of_route <= 0:
                raise ValueError
            else:
                return length_of_route
        except Exception:
            print("Значенте введено неверно, поробуйте снова, пример (10 км)")
    
def fuel_cons():
    return fuel_input() * length_of_route_input()

print(f"Необходимое топливо: {fuel_cons()}")
```
