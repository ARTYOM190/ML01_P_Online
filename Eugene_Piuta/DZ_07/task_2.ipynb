#Задача: 
Есть данные, что тест на болезнь дает результат с вероятностью в 98%. Число больных в популяции у нас 1%. 
Вопрос - с какой вероятностью положительный тест будет положительным при случайном тестировании. 

Точность теста (T): 0.98(вероятность того, что тест будет положительным при наличии болезни).
Вероятность заболеть (Pз): 0.01 (доля больных ).
Вероятность не заболеть (Pнз): 1 - 0.01 = 0.99.
Вероятность ложноположительного результата (Pп): 1-0.98 = 0.02.
Применяя теорему Байеса, получаем вероятность того, что человек действительно болен при положительном результате теста: 
P = T*Pз/Pо
Pо - общая вероятность получения положительного результата
Po = T*Pз + Pп*Pнз = 0.98*0.01+0.02*0.99 = 0.0296
P = T*Pз/Pо = 0,98*0,01/0,0296 = 0,331


#Вывод:
Вероятность того, что человек действительно болен при положительном результате теста составляет 33,1%.
